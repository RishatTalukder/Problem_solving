# Introduction to Trees

## What is a Tree?

A `tree` is a `non-linear` data structure that is used to represent `hierarchical data`. Most of the data structures that we have studied so far like `arrays`, `linked lists`, `stacks`, `queues`, etc. are `linear data structures`. A `linear data structure` is a `data structure` in which the `elements` are `arranged` in a `linear order`. A `non-linear data structure` is a `data structure` in which the `elements` are `arranged` in a `spread out` manner. 

Most of the time a `tree` is represented as a `Linked List` with `TWO` or `MORE` `pointers` at each `node`.

There are other ways to represent a `tree` as well. For example, a `tree` can be represented as an `array` or a `matrix` as well.

But we will be using the `Linked List` representation of a `tree` in this article and mainly `binary trees` and `binary search trees`.

## Basic Terminologies

Before we start learning about `trees`, let's learn about some of the `basic terminologies` that are used in `trees`.

*** Node

Every `element` in a `tree` is called a `node`. A `node` is a `data structure` that contains some `data` and `pointers` to other `nodes`. A `node` can be represented as a `class` in `python` or a `struct` in `C/C++`.

*** Root

The `topmost node` in a `tree` is called the `root` of the `tree`. A `tree` can have `only one root`.

![Root](https://media.geeksforgeeks.org/wp-content/uploads/20221107210238/Capture1-660x440.JPG)

*** Parent

A `node` that has `child nodes` is called a `parent node`. A `node` can be a `parent` to `one or more child nodes`.

The root node is the `parent` of all the `nodes` in the `tree`.

*** Child

A `node` that has a `parent node` is called a `child node`. A `node` can be a `child` to `only one parent node`.

*** Siblings

`Nodes` that have the `same parent node` are called `siblings`.

*** Leaf

A `node` that does not have any `child node` is called a `leaf node` or a `leaf`. A `leaf node` is also called an `external node`.

![Leaf](https://afteracademy.com/images/what-is-a-tree-data-structure-terminology-1-157b14ad63d64878.png)

*** FULL TREE

A `tree` in which `every node` has `zero or two child nodes` is called a `full tree`.

*** COMPLETE TREE

A `tree` in which `the last level` is `filled` from `left to right` is called a `complete tree`.

*** PERFECT TREE

A `tree` in which `all the internal nodes` have `two child nodes` and `all the leaf nodes` are at the `same level` is called a `perfect tree`.

![FULL VS COMPLETE VS PERFECT](https://study.com/cimages/multimages/16/0e0646ba-30e5-40d9-b45c-a138f038f05b_full_complete_perfect.png)

These are the `basic terminologies` that are used in `trees`. Now let's learn about `binary trees`.

# Binary Tree

A `binary tree` is a `tree` in which `every node` has `zero, one, or two child nodes`. A `binary tree` can be `empty` as well. A `binary tree` can be represented as a `Linked List` with `two pointers` at each `node`.




A `binary tree` can be `full`, `complete`, or `perfect` as well.

> Binary tree is a tree in which every node has zero, one, or two child nodes.


## Binary Tree Traversal


A `binary tree` can be represented as a `Linked List` with `two pointers` at each `node`. The `first pointer` points to the `left child node` and the `second pointer` points to the `right child node` as shown on the picture above.


`Tree traversal` is the process of `visiting` or `updating` each `node` in a `tree`. There are `two ways` to `traverse` a `tree`:

*** Depth First Traversal

In `depth first traversal`, we `visit` the `nodes` in a `tree` `depth wise`. There are `three ways` to `traverse` a `tree` in `depth first traversal`:

1. Preorder Traversal
2. Inorder Traversal
3. Postorder Traversal

*** Breadth First Traversal

In `breadth first traversal`, we `visit` the `nodes` in a `tree` `level wise`. We `visit` the `nodes` from `left to right` in `breadth first traversal`.
